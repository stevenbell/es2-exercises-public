In this project, we're going to create another simulation, this time to model a physical scenario.
Consider the pendulum below, which should be familiar if you've taken physics:



The only forces acting on the pendulum mass are <span style="color: #a40000">gravity</span> (straight down) and the tension of the string.
The string tension will cancel out the <span style="color: #4e9a06">component of gravity parallel to the string</span>, leaving only the <span style="color: #f57900">component of gravity perpendicular to the string</span> to act on the pendulum.

Assuming that positive motion is to the right, this component of the force is given by 
$$F = -mg\sin\theta$$
where $g$ is the acceleration of gravity (9.81 $\mathrm{m}/\mathrm{s^2}$), and $\theta$ is the angle relative to vertical.

This is the point where the physics texts all wave their hands and say that if $\theta$ is small, we can just write $$F = -mg\theta$$ because $\theta \approx \sin\theta$ for "small" values of $\theta$.  But what counts as "small"?  And what happens if $\theta$ isn't small?

The math gets pretty challenging if you try to solve this analytically, which is why physics texts make the small-angle approximation and get on with things.  (You can read the full derivation [on Wikipedia](https://en.wikipedia.org/wiki/Pendulum_(mechanics)).)

We're going to take a different approach using Python:

When we're done, you'll have a simulation that didn't make the small-angle approximation and didn't require any calculus.  Win-win.
