## Reading a file

To read the contents of a file, we first have to open it:

```python
file = open("lecture12_testfile.txt")
```

Once we have an open file, we can start reading from it:

```python
result = file.read(5)
print(result)
```

A very common thing to do is to read a file one line at a time in a `for` loop:

## Writing files

Writing files follows basically the same sequence as reading files:
1. Open the file with `open()`
2. Write data to the file with `write()`
3. Close the file with `close()`

There's one catch: files are opened read-only by default, so if you want to open a file for writing, you need to specify a second parameter in the `open()` call:

```python
outfile = open("outfile.txt", "w") # The "w" specifies that we're writing a new file
outfile.write("Woohoo, this file was generated by Python!")
outfile.close()
```
